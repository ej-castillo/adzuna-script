```python
import requests
import json
import os
import pandas
```

```python
app_id = 'ffd2872c'
api_key = 'a96440d14c2aae542bc5306da5076874'
```

```python
state = 'New York'
county = 'New York City'
city = 'Manhattan'
num_jobs = 10
```

```python
state_f = '&location1=' + state.replace(' ', '%20') if not pandas.isnull(state) else ''
county_f = '&location2=' + county.replace(' ', '%20') if not pandas.isnull(county) else ''
city_f = '&location3=' + city.replace(' ', '%20') if not pandas.isnull(city) else ''

req_url = f'http://api.adzuna.com/v1/api/jobs/us/search/1?app_id={app_id}&app_key={api_key}&location0=us{state_f}{county_f}{city_f}&results_per_page={num_jobs}&content-type=application/json'

req_url
```

```python
req_ret = requests.get(req_url)
```

```python
r = req_ret.json()
results = r["results"]
```

```python
for index, result in enumerate(results):
    if index == 1:
        print(result['description'])
    result['company'] = result['company']['display_name']
    result['category'] = result['category']['label']
    result['state'] = 'California'
    result['city'] = 'Los Angeles'
```

```python
pandas.set_option('display.max_columns', None)
pandas.set_option('display.max_rows', None)

df = pandas.DataFrame(results)
df = df[['id', 'created', 'company', 'title', 'category', 'description', 'city', 'state', 'latitude', 'longitude', 'contract_type', 'contract_time', 'redirect_url']]
df
```
